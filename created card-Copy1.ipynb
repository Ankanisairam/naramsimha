# using the Elbow method in clustering

```python
# Step 1: Import necessary libraries
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Step 2: Generate or load data
# For demonstration, let's create some synthetic data
X, _ = make_blobs(n_samples=300, centers=2, cluster_std=0.60, random_state=0)

# Step 3: Calculate WCSS for different numbers of clusters
wcss = []
for i in range(1, 11):  # Number of clusters from 1 to 10
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)  # Inertia is another term for WCSS

# Step 4: Plot the elbow graph
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# Data visualization in clustering 

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
x = np.random.rand(2000,2)
```

```python
kmeans = KMeans(n_clusters = 2,init = 'k-means++',random_state=42)
y_predict = kmeans.fit_predict(x)
```

```python
plt.scatter(x[y_predict ==0,0], x[y_predict == 0,1],s=50,c='blue',label='class 0')
plt.scatter(x[y_predict ==1,0], x[y_predict ==1,1],s=50,c='green',label ='class 1')
# mtp.scatter(x[y_predict ==2,0], x[y_predict == 2,1],s=50,c='red',label= 'class 2')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200, c= 'yellow', marker ='x', label= 'centroids')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.legend()
plt.show()
        
```
